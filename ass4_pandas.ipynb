### 11. How many students got A, B, C, F?

```python
dictionery={'id':[1,2,3,4,5],'marks':[12,78,99,44,90]}
data=pd.DataFrame(dictionery)
data_marks=dictionery['marks']
grade=list()
c=0
a=0
f=0
b=0
for x in data_marks:
    if x>=50 and x<60:
        g='C'
        grade.append(g)
        c+=1
    elif x>=60 and x<70:
        g='B'
        grade.append(g)
        b+=1
    elif x>=70:
        g='A'
        grade.append(g)
        a+=1
    else:
        g='F'
        grade.append(g)
        f+=1
print(grade)
print(a,"-->A  ",b,"-->B  ",c,"-->C",f,"-->F")
```
